```python
# data analysis and wrangling
import numpy as np
import pandas as pd
import json

# data standardization
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# deep learning
from keras import models
from keras import layers
from tensorflow.keras.utils import to_categorical
```

```python
from pycaret.classification import *
```

```python
from pycaret.utils import check_metric
```

```python
csv = pd.read_csv('preprocessed_train.csv', encoding='euc-kr')

train = csv[:8000]
test = csv[8000:]
```

```python
train
```

```python
exp1 = setup(train, target = 'label',
             ignore_features=['분석데이터'])
```

```python
# ML 모델 성능 비교
top5 = compare_models(sort='Accuracy', n_select=5)
```

```python
# 모델 튜닝, blending
tuned_top5 = [tune_model(i) for i in top5]
blender_top5 = blend_models(estimator_list=tuned_top5)
```

```python
# 모델 최종 확정, 전체 데이터 학습
final_model = finalize_model(blender_top5)
```

```python
# 테스트셋 predict
prediction = predict_model(final_model, data=test)
```

```python
prediction
```

```python
# metric 확인
check_metric(prediction['label'], prediction['Label'], metric = 'Accuracy')
```
