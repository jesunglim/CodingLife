```python
from datetime import datetime, timedelta

def date_range(start, end):
    start = datetime.strptime(start, "%Y-%m-%d")
    end = datetime.strptime(end, "%Y-%m-%d")
    dates = [(start + timedelta(days=i)).strftime("%Y-%m-%d") for i in range((end-start).days+1)]
    return dates

dates = date_range("2018-08-02", "2022-03-09")
print(dates)
```

```python
len(dates)
```

```python
for date in dates:
    for i in range(13):
        to = date + " " + str(i) + ':00'
        count = 60
        df2 = pyupbit.get_ohlcv(ticker=ticker,interval=interval,to=to,count=count)
        df = pd.concat([df,df2])
```
