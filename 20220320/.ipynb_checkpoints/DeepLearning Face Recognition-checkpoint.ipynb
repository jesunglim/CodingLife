```python
import numpy as np
import matplotlib.pyplot as plt

from deepface import DeepFace
```

# Stage 1 and 2 : Detection and Alignment

### 동일인물의 사진으로 준비한 후, 아래의 함수를 실행시키는 것만으로 
### Detect와 Alignment된 사진을 얻을 수 있습니다(사람의 이미지라면) 

```python
shin1 = DeepFace.detectFace("shin1.jpg")
shin2 = DeepFace.detectFace("shin2.jpg")
```

```python
plt.imshow(shin1)
```

```python
plt.imshow(shin2)
```

# Stage 3 : Representation

#### Deep Learning은 이 Representation단계부터 사용됩니다. 
#### 먼저 얼굴의 이미지를 CNN에 통과시켜 줍니다
#### 하지만 아직 사람의 얼굴을 분류(Classification)하는 단계는 아닙니다. 
#### Alignment된 얼굴 이미지를 CNN에 통과시키는 이유는 embedding을 하기 위해서입니다. 

```python
from deepface.basemodels import VGGFace, OpenFace, Facenet, FbDeepFace
model = VGGFace.loadModel()
# model = Facenet.loadModel()
# model = OpenFace.loadModel()
# model = FbDeepFace.loadModel()
```

#### 지금까지 언급한 얼굴 인식모델은 대용량의 데이터셋으로 학습된 Pre_trained model이 존재
#### 이미지의 Representation를 출력하는 것이 최종 Output

## ArcFace로 embedding

```python
from arcface import ArcFace
face_rec = ArcFace.ArcFace()
```

```python
shin1_representation = face_rec.calc_emb(shin1)
shin2_representation = face_rec.calc_emb(shin2)
```

### 현재 단계에서 얻은 Representation은 1D 행렬 즉 한 줄
#### 시각화를 위해 이 행렬을 2D 행렬로 변경 
###### ( for문으로 200개의 같은 행렬 쌓기)

```python
shin1_graph = []; shin2_graph = []
 
for i in range(0, 200):
    shin1_graph.append(shin1_representation)
    shin2_graph.append(shin2_representation)
    
shin1_graph = np.array(shin1_graph)
shin2_graph = np.array(shin2_graph)
```

```python
fig = plt.figure()
 
ax1 = fig.add_subplot(3,2,1)
plt.imshow(shin1)
plt.axis('off')
 
ax2 = fig.add_subplot(3,2,2)
im = plt.imshow(shin1_graph, interpolation='nearest', cmap=plt.cm.ocean)
plt.colorbar()
 
ax3 = fig.add_subplot(3,2,3)
plt.imshow(shin2)
plt.axis('off')
 
ax4 = fig.add_subplot(3,2,4)
im = plt.imshow(shin2_graph, interpolation='nearest', cmap=plt.cm.ocean)
plt.colorbar()
 
plt.show()
```

# Verification

#### 두 개의 Representation이 같은지 다른지 비교하는 방법은
#### 둘 사이의 euclidean distance(유클리드 거리)를 구하면 됨

```python
# shin1의 vector 값에서 shin2의 vector값을 뺀 후 제곱
distance_vector = np.square(shin1_representation - shin2_representation)
# 제곱한 값에 루트를 씌움
distance = np.sqrt(distance_vector.sum())
```

```python
distance_graph = []
for i in range(0, 200):
    distance_graph.append(distance_vector)
distance_graph = np.array(distance_graph)
    
ax6 = fig.add_subplot(3,2,6)
im = plt.imshow(distance_graph, interpolation='nearest', cmap=plt.cm.ocean)
plt.colorbar()
```

```python
if distance < 0.55:
    print("Distance: " + str(distance))
    print("TRUE")
else:
    print("Distance: " + str(distance))
    print("FALSE")
```
