```python
cnt = 0

for i in range(len(csv)):
    for contry in contry_lst:
        if csv.loc[i]['LOCATION'] == contry:
            df[contry][cnt] = csv.loc[i]['Value']
            cnt += 1
    if cnt >= 5:
        cnt = 0
```
